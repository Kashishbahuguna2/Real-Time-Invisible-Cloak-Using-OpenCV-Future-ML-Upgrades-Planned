```python
import cv2
import numpy as np 
import time 

# step 1 : capatured background 

captured  = cv2.VideoCapture(0) #open webcam 
time.sleep(2) # give time to camera to adjust 

background = 0 

# capturing the background frame

for i in range(50):
    ret , background = captured.read()
    if not ret:
        continue 

#fliping the background 

background = np.flip(background , axis = 1)

print("background captured successfully!!")


# step 2: invisiblity effect loop 

while captured.isOpened():

    ret , frame  = captured.read()
    if not ret :
        break 

    #fliping the frame 

    frame = np.flip(frame , axis= 1)

    # converting the color from BGR to HSV

    hsv = cv2.cvtColor(frame , cv2.COLOR_BGR2HSV)

    #step 3 creating mask for red cloak 
    # red color has two hsv range 

    lower_red1 = np.array([0 , 120 , 70])
    upper_red1 = np.array([10 , 255 , 255])

    lower_red2 = np.array([170 , 120 , 70])
    upper_red2 = np.array([180 , 255 , 255])

    #creating masks

    mask1 = cv2.inRange(hsv , lower_red1 , upper_red1)
    mask2 = cv2.inRange(hsv , lower_red2 , upper_red2)

    # final mask

    mask = mask1 + mask2 

    # refining the mask

    mask = cv2.morphologyEx(mask , cv2.MORPH_OPEN ,np.ones((3,3) , np.uint8)) # remove noise 

    mask = cv2.dilate(mask , np.ones((3,3) , np.uint8)) # enlarge mask area and give cloak area clear 

    #step 4: create invisible effect (meams clock wali jagh background se replacekarna)

    cloak_hidden = cv2.bitwise_and(background , background , mask =  mask ) #Cloak wali jagah background se replace karo

    mask_inverse = cv2.bitwise_not(mask) #Cloak nahi hai waha normal frame rakhna

    normal_part = cv2.bitwise_and(frame , frame , mask = mask_inverse ) # Frame ka sirf wo part rakhta hai jaha cloak nahi ha

    final_output = cloak_hidden + normal_part # final image combine karna

    cv2.imshow("invisibility cloak - press q to exits " , final_output)

    if cv2.waitKey(1) == ord('q'):
        break 

captured.release()
cv2.destroyAllWindows()










    

```
